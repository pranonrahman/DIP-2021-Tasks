## Solution to task 1
In this task, we need to implement a thresholding operation. As we know, for a gray-scale image, the pixel values shows the brightness intensity of that pixel. If the pixel value is $255$ then it is white and if the pixel value is $0$ then the pixel is black. In the thresholding function, we needed to make all a pixel value $255$ if the pixel value is higher than a threshold value. Otherwise, we will have to set the pixel value $0$.  \\
The pseudocode of our task is:
``` python 
for pixel in pixels:
    if img[pixel] > threshold:
        img[pixel] = 255
    else:
        img[pixel] = 0
```
Now, we will implement the function. In this section we could simply use a nested loop to implement the function. 

````python
def threshold(image, threshold_value = 135):
    data = np.array(image)
    #data.setflags(write=1)
    #print(f'Shape of the image is: {data.shape}')
    rows,cols = data.shape
    for row in range(0,rows):
        for pixel in range(0,cols):
            if data[row][pixel] <= threshold_value:
                data[row][pixel] = 0.0
            else:
                data[row][pixel] = 255
    image = Image.fromarray(data)
    return image
````

````python
image = Image.open('/content/Fig3.09(a).jpg')
display(image)



binary_image = threshold(image,210)


binary_image.save('darker_city.jpg')
display(binary_image)
````

### Discussion

In this figure first we can see a given image. The image seems a bit washed out. But as we ran the thresholding operation, we simply eliminated the gray values and made the image binary. The pixel value is either $0$ (black) or $255$ (white). Now, as we can see from the function the image was selected the thresholding value to 210. So all intensity value less than or equal to $210$ was set to black. Thus, we can have a better more detailed image. 

Now, if we increase or decrease the threshold value, we will be able to see the difference. First we set threshold value to $100$ and then $150$ then $200$ and at last $250$

````python
image = Image.open('/content/Fig3.09(a).jpg')

threshold_100 = threshold(image,100)
threshold_150 = threshold(image,150)
threshold_200 = threshold(image,200)
threshold_250 = threshold(image,250)

display(threshold_100)
print("image with threshold 100")
display(threshold_150)
print("image with threshold 150")
display(threshold_200)
print("image with threshold 200")
display(threshold_250)
print("image with threshold 250")
````

As we can see, if we reduce the threshold value, the image will be mostly washed out as a slightly bright pixel value will have intensity of $255$. But as we increase the threshold value, the image gets darker and darker.

## Task 2
In this task we are asked to build a histogram for a given image. We will use matplotlib to display the histogram. 

First we will need to get the frequency for each intensity values of the image. Then, we can simply use pyplot to print it as a scatterplot. 

````python
def create_histogram(image, title = 'Histogram of given image'):
    data = np.array(image)

    frequency = np.zeros(256)

    rows,cols = data.shape
    for row in range(0,rows):
        for pixel in range(0,cols):
            frequency[data[row][pixel]] += 1

    plt.figure(figsize=[10,8])

    plt.bar(range(256), frequency, width = 0.5, color='#0504aa',alpha=0.7)
    plt.xlabel('Intensity',fontsize=15)
    plt.ylabel('Frequency',fontsize=15)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.ylabel('Frequency',fontsize=15)
    plt.title(title,fontsize=15)
    plt.show()
````

````python
create_histogram(image, 'Histogram of Given Image')
print('')
create_histogram(binary_image, 'Histogram of Binary Image')
````

In this figure we generated two histogram. One for the given image on the beginning. Here, we can see that the image has a lot of bright pixel. That's why the image seems a bit washed out. But, after we ran the thresholding function we have a lot of pixel valued $0$. And a less number of pixels values 255. But we don't have any values in between $0$ and $255$
